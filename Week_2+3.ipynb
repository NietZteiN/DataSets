```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns for outlier detection
numerical_cols = ['KM Travelled', 'Price Charged', 'Cost of Trip']

# Create box plots for each numerical column
for col in numerical_cols:
  plt.figure()
  sns.boxplot(x=combined_df[col])
  plt.title(f'Box Plot of {col}')
  plt.show()
```

```python
import numpy as np

# Calculate the third quartile for 'Price Charged'
q3 = combined_df['Price Charged'].quantile(0.75)

# Replace upper outliers with the third quartile
combined_df['Price Charged'] = np.where(combined_df['Price Charged'] > q3, q3, combined_df['Price Charged'])

# Verify the changes (optional)
sns.boxplot(x=combined_df['Price Charged'])
plt.title('Box Plot of Price Charged (After Outlier Handling)')
plt.show()
```

Seasonality in Cab Usage: Cab usage shows significant seasonality, with peaks during certain times of the year (e.g., holidays, summer).



```python
# Extract the year from the 'Date of Travel' column
combined_df['Year'] = combined_df['Date of Travel'].dt.year

# Group the data by year and count the number of trips
trips_by_year = combined_df.groupby('Year')['Transaction ID'].count()

# Create a bar chart of cab usage by year
plt.figure()
trips_by_year.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Trips')
plt.title('Cab Usage by Year')
plt.show()
```

```python
# Define a function to map months to seasons
def get_season(month):
  if month in [12, 1, 2]:
    return 'Winter'
  elif month in [3, 4, 5]:
    return 'Spring'
  elif month in [6, 7, 8]:
    return 'Summer'
  else:
    return 'Fall'

# Extract the month from the 'Date of Travel' column and apply the season mapping
combined_df['Month'] = combined_df['Date of Travel'].dt.month
combined_df['Season'] = combined_df['Month'].apply(get_season)

# Group the data by season and count the number of trips
trips_by_season = combined_df.groupby('Season')['Transaction ID'].count()

# Create a bar chart of cab usage by season
plt.figure()
trips_by_season.plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('Number of Trips')
plt.title('Cab Usage by Season')
plt.show()
```

Hypothesis: there is a difference in revenue per ride depending on the population of the city

```python
# Calculate revenue per ride for each city
revenue_per_ride = combined_df.groupby('City')['Price Charged'].sum() / combined_df.groupby('City')['Transaction ID'].count()

# Extract population data from df2
city_populations = df2.set_index('City')['Population']

# Combine revenue per ride and population data
revenue_population_df = pd.concat([revenue_per_ride, city_populations], axis=1)
revenue_population_df.columns = ['Revenue per Ride', 'Population']

# Create a scatter plot
plt.figure()
plt.scatter(revenue_population_df['Population'], revenue_population_df['Revenue per Ride'])
plt.xlabel('Population')
plt.ylabel('Revenue per Ride')
plt.title('Revenue per Ride vs. Population')
plt.show()
```

Hypothesis: dfferent age groups will show different amounts of cab usage

```python
# Group the data by age and count the number of trips
trips_by_age = combined_df.groupby('Age')['Transaction ID'].count()

# Create a bar chart of cab usage by age
plt.figure()
trips_by_age.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Number of Trips')
plt.title('Cab Usage by Age')
plt.show()
```

```python
# Calculate the average number of trips per customer
average_trips_per_customer = combined_df.groupby('Customer ID')['Transaction ID'].count().mean()

# Print the result
print("Average number of trips per customer:", average_trips_per_customer)
```

```python
# Create age brackets
combined_df['Age Bracket'] = pd.cut(combined_df['Age'], bins=range(0, 100, 10), right=False)

# Calculate average trips per customer by age bracket
average_trips_by_age_bracket = combined_df.groupby('Age Bracket')['Transaction ID'].count() / combined_df.groupby('Age Bracket')['Customer ID'].nunique()

# Create a bar chart
plt.figure()
average_trips_by_age_bracket.plot(kind='bar')
plt.xlabel('Age Bracket')
plt.ylabel('Average Number of Trips per Customer')
plt.title('Average Cab Usage by Age Bracket')
plt.show()
```

Hypothesis: cab usage will differ depending on the custsomer's income bracket

```python
# Create income brackets
combined_df['Income Bracket'] = pd.cut(combined_df['Income (USD/Month)'], bins=range(0, 150000, 10000), right=False)

# Calculate average trips per customer by income bracket
average_trips_by_income_bracket = combined_df.groupby('Income Bracket')['Transaction ID'].count() / combined_df.groupby('Income Bracket')['Customer ID'].nunique()

# Create a bar chart
plt.figure()
average_trips_by_income_bracket.plot(kind='bar')
plt.xlabel('Income Bracket')
plt.ylabel('Average Number of Trips per Customer')
plt.title('Average Cab Usage by Income Bracket')
plt.show()
```
